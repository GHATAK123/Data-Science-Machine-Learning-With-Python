```python
X_Boston_Data = Boston_Data.data/Boston_Data.data.max()
X_Boston_Data
```

```python
y_Boston_Data = Boston_Data.target
```

```python
n_samples = len(X_Boston_Data)
```

```python
# Training data of both. x and y
X_train = X_Boston_Data[:int(0.9*n_samples)]
X_train
```

```python
y_train = y_Boston_Data[:int(0.9*n_samples)]
y_train
```

```python
# Testing data for both x and y
X_test = X_Boston_Data[:int(0.9*n_samples)]
X_test
```

```python
y_test = y_Boston_Data[:int(0.9*n_samples)]
y_test
```

```python
# Data Modeling
Knn = neighbors.KNeighborsClassifier()
linear = linear_model.LinearRegression() #leghthbredthfiguresize=lbfgs
linear
```
